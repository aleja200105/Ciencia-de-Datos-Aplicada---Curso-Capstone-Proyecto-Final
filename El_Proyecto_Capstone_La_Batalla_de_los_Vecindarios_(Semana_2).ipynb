```python
import pandas as pd
import numpy as np
```

```python
#Vuelvo a poner los pasos para obtener el dataframe de la primera parte del proyecto
df = pd.read_html('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1008658788')
df_code = df[0]
df_code2 = df_code[df_code['Borough']!= 'Not assigned']
df_code2 = df_code2.reset_index(drop=True)
df_code2.head(12)
```

```python
#Datos que tengo que buscar (los de codigo postal de Toronto)
postal_code = df_code['Postal Code']
postal_code
```

```python
df_toronto_coord = pd.read_csv('Geospatial_Coordinates.csv')
print(df_toronto_coord.head())
print(df_toronto_coord.shape)
```

```python
#Así queda el dataframe con las latitudes y longitud
df_toronto_coord.head(12)
```

```python
#Ahora uno el datframe de codigos postales y este último de coordenadas
df_toronto_final = pd.merge(df_code,df_toronto_coord)
df_toronto_final.head(12)
```

```python
# Importación del paquete de geopy para determinar las coordenadas la ciudad de Toronto
!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim
```

```python
# Buscamos las coordenadas para realizar el mapa
address = 'Toronto, Ontario'
geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('Las coordenadas de '+address,' son {}, {}.'.format(latitude,longitude))
```

```python
#Generación de un mapa con los barrios superpuestos de Toronto
# Importamos el paquete Folium para realizar el mapa
import folium
from IPython.core.display import display, HTML

def folium_deepnote_show(m):
    data = m.get_root().render()
    data_fixed_height = data.replace('width: 100%;height: 100%', 'width: 100%').replace('height: 100.0%;', 'height: 609px;', 1)
    display(HTML(data_fixed_height))
```

```python
# Gráfico del mapa de la ciudad de Toronto
map_toronto = folium.Map(location=[latitude, longitude])
folium_deepnote_show(map_toronto)
```

```python

# Añadimos marcadores de los municipios y vecindarios
for lat,lng, borough, neighborhood in zip(df_toronto_final['Latitude'],df_toronto_final['Longitude'],df_toronto_final['Borough'],df_toronto_final['Neighbourhood']):
    label= 'Neighborhood {}, Borough {}'.format(neighborhood,borough)
    label= folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup= label,
        color= 'red',
        fill=True,
        fill_color='yellow',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)

folium_deepnote_show(map_toronto)
```

```python
# Numero de vecindarios por municipios
df_toronto_borough = pd.value_counts(df_toronto_final['Borough'])
df_toronto_borough = pd.DataFrame(df_toronto_borough).reset_index()
df_toronto_borough = df_toronto_borough.rename(columns={'index':'Ciudad','Borough':'Número de vecindarios'})
df_toronto_borough
```
